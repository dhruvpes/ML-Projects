```python
df = pd.read_csv("spam.csv")
df.head()
```

```python
df = df.drop(['Unnamed: 2', 'Unnamed: 3', 'Unnamed: 4'], axis = 1)
```

```python
df.head(10)
```

```python
df.groupby('v1').describe()
```

```python
df.groupby('v2').describe()
```

```python
df['spam'] = df['v1'].apply(lambda x: 1 if x == 'spam' else 0)
df.head(15)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df['v2'], df['spam'], test_size=0.25)
```

```python
## Now proceeding to make use of COunt vectorizer (try to use word2vec, bert also)
from sklearn.feature_extraction.text import CountVectorizer
v = CountVectorizer()
X_train_count = v.fit_transform(X_train.values)
X_train_count.toarray()[:3]


```

```python
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
model.fit(X_train_count, y_train)
```

```python
emails = ["Hey Dhruv, it was great meeting you today. I hope we can discuss further and plan on what songs we woud like to perform for our College fest. Regards",
          "Free Parking at new mall!! Tap this link to avail full discount and get a chance to win a Dubai trip for 10 days.. absolutely 0 cost!!"
         ]
```

```python
emails[1]
```

```python
emails_count = v.transform(emails)
model.predict(emails_count)
```

```python
X_test_count = v.transform(X_test)
model.score(X_test_count, y_test)
```

```python
from sklearn.pipeline import Pipeline
```

```python
clf = Pipeline([('vectorizer', CountVectorizer()), 
               ('nb', MultinomialNB())
               ])
```

```python
clf.fit(X_train, y_train)
clf.score(X_test, y_test)
```
